## Question 5

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0., 2. * np.pi, 101)
z = np.exp(1.j * theta)
zeta = z + 1. / z

plt.plot(np.real(zeta), np.imag(zeta))
plt.axis('equal')
plt.show()
```

```python
z = 2. * np.exp(1.j * theta)
zeta = z + 1. / z

plt.plot(np.real(zeta), np.imag(zeta))
plt.axis('equal')
plt.show()
```

```python
maximum = 5.
for radius in np.linspace(0.,maximum,11):
    z = radius * np.exp(1.j * theta)
    zeta = z + 1. / z
    colour_unit = radius/maximum

    plt.plot(np.real(zeta), np.imag(zeta), label=round(radius,2), 
             color = (colour_unit,0,1-colour_unit))

plt.legend()
plt.axis('equal')
plt.show()
```

## Question 6

```python
h0 = 10.; a = 0.25; x0 = 2.; b = 4.; y0 = -1.

x = np.linspace(-10., 15., 26)
y = np.linspace(-5., 5., 11.)
xmat = np.zeros((x.size, y.size))
ymat = np.zeros((x.size, y.size))

for i in range(x.size):
    for j in range(y.size):
        xmat[i, j] = x[i]
        ymat[i, j] = y[j]

h = h0 - a * (xmat - x0)**2 - b * (ymat - y0)**2     
```

```python
plt.plot(x, h[:, 6]); plt.show()
```

```python
plt.plot(y, h[11, :]); plt.show()
```

```python
plt.contour(xmat, ymat, h, [-50,-10,-1,0,2,7]); plt.show()
```

```python
plt.imshow(h)
```

## Question 8

```python
r = np.linspace(0., 5., 51)
theta = np.linspace(0., 2. * np.pi, 101)

xmat = np.zeros((r.size, theta.size))
ymat = np.zeros((r.size, theta.size))
psi = np.zeros((r.size, theta.size))

for ir in range(r.size):
    for ith in range(theta.size):
        xmat[ir, ith] = r[ir] * np.cos(theta[ith])
        ymat[ir, ith] = r[ir] * np.sin(theta[ith])
        psi[ir, ith] = r[ir] * np.exp(-r[ir]) * np.cos(theta[ith])**2

plt.contour(xmat, ymat, psi)
plt.show()
```

```python
maximum = 0
coordinates = (0,0)

for ir in range(r.size):
    for ith in range(theta.size):
        if psi[ir,ith] > maximum:
            maximum = psi[ir,ith]
            coordinates = (r[ir],theta[ith])
```

## Question 9

```python
x = np.linspace(-17., 22., 40)
y = np.linspace(-10., 15., 26)
xmat = np.zeros((x.size, y.size))
ymat = np.zeros((x.size, y.size))

for i in range(x.size):
    for j in range(y.size):
        xmat[i, j] = x[i]
        ymat[i, j] = y[j]
        
h = xmat**2 - ymat**2 + 2*xmat*ymat*1j
```

```python
plt.contour(xmat, ymat, np.real(h), colors = 'red')
plt.contour(xmat, ymat, np.imag(h))
plt.axis('equal')

plt.show()
```
